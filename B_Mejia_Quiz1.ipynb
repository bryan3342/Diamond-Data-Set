**<h2> What proportion of diamonds are between .30 and 1.08 carats?</h2>**

```python
# @ Inference 1

# Create data frame with conditionals to satisfy problem to parse data without including invalid data.
filtered_df = df[(df['Carat'] >= 0.30) & (df['Carat'] <= 1.08) & (df['x'] > 0) & (df['y'] > 0) & (df['z'] > 0)]

# Used pandas library drop_duplicates() function to not have to convert from list to set back to list
unique_x_dimensions = filtered_df['x'].drop_duplicates().tolist()
unique_y_dimensions = filtered_df['y'].drop_duplicates().tolist()
unique_z_dimensions = filtered_df['z'].drop_duplicates().tolist()

data_x_table = pd.DataFrame({'Unique x Dimensions': unique_x_dimensions})
data_y_table = pd.DataFrame({'Unique y Dimensions': unique_y_dimensions})
data_z_table = pd.DataFrame({'Unique z Dimensions': unique_z_dimensions})

# Concatenate all 3 lists into a singular data frame for simpler and robust presentation.
combined_dimensions_table = pd.concat([data_x_table, data_y_table, data_z_table], axis=1)
display(combined_dimensions_table)

plt.figure(figsize=(10, 6))
plt.subplot(1, 3, 1)
plt.hist(data_x_table, bins=20, alpha=0.5, label='x Dimensions')
plt.hist(data_y_table, bins=20, alpha=0.5, label='y Dimensions')
plt.hist(data_z_table, bins=20, alpha=0.5, label='z Dimensions')

plt.xlabel('Dimensions')
plt.ylabel('Frequency')
plt.title('Distribution of Dimensions')

plt.legend()
plt.grid(True)
plt.show()
```

**<h2>How many diamonds have equal x and y dimensions?</h2>**

```python
# @ Inference 2
# Create a list to collect integer location of filtered diamonds
filtered_diamonds = []

for i in range(len(df['x'])):
  if df['x'][i] == df['y'][i]:
    filtered_diamonds.append(df.iloc[i]) # Getting information based on integer location

filtered_df = pd.DataFrame(filtered_diamonds)

display(filtered_df)
print(f'There are a total of {len(filtered_df)} diamonds with the same x and y dimensions.')
```

**<h2>How many diamonds have a Premium cut or Better? Note that the possible values for the quality of a cut are ranked in ascending order as follows : Fair / Good / Very Good / Premium / Ideal</h2>**

```python
# Inference 4
# Using conditionals, store all diamonds that satisfy conditionals through comparisons.

premium_or_better_cut = []

for cut in df['Cut']:
  if cut == 'Premium' or cut == 'Ideal':
    premium_or_better_cut.append(cut)

print(f'There are a total of {len(premium_or_better_cut):,} diamonds with a Premium cut or Better.')
```

**<h2>Which diamond has the highest price per carat? What is its value?</h2>**

```python
# @ Inference 5
# Convert Data Frame columns to lists to simplify traversal

highest_price_per_carat = 0
diamond_index = 0
Price = df['Price'].to_list() # Convert data frame columns to lists to make it easier to access
Carat = df['Carat'].to_list()

for i in range(len(Price)): # Iterate through entire Price list
  if Price[i] / Carat[i] > highest_price_per_carat: # Compare the current index for Price and Carat and compare to current
    highest_price_per_carat = Price[i] / Carat[i]
    diamond_index = i

print(f"The diamond that has the highest per carat is {diamond_index:,} and it's value is {highest_price_per_carat:.2f}")
```
